```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

# Opening Images in Notebook

```python
pic = cv2.imread('DATA/00-puppy.jpg')
print(type(pic))
plt.imshow(pic)
```

> Matplotlib -> RGB

> OpenCv -> BGR

```python
pic = cv2.imread('DATA/00-puppy.jpg')
pic = cv2.cvtColor(pic, cv2.COLOR_BGR2RGB)
plt.imshow(pic)
```

# Grayscaling

```python
pic_gray = cv2.imread('DATA/00-puppy.jpg', cv2.IMREAD_GRAYSCALE)
print(pic_gray.shape)
plt.imshow(pic_gray, cmap='gray')
```

# Resizing

```python
pic_resized = cv2.resize(pic, (975,650)) # w, h
print(pic_resized.shape) # r, c (h, w)
plt.imshow(pic_resized)
```

# Flipping

```python
plt.imshow(cv2.flip(pic, 0))
```

```python
plt.imshow(cv2.flip(pic, 1))
```

```python
plt.imshow(cv2.flip(pic, -1))
```

# Custom Axis

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.imshow(pic)
```
