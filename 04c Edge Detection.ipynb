```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('DATA/sammy_face.jpg')
plt.imshow(img)
```

**CHOOSING THRESHOLDS**

```python
med_val = np.median(img) 
lower = int(max(0, 0.7* med_val))
print(lower)
upper = int(min(255,1.3 * med_val))
print(upper)
```

# Canny Edge Detection

```python
edges = cv2.Canny(image=img, threshold1=lower , threshold2=upper)
plt.imshow(edges)
```

```python
edges = cv2.Canny(image=img, threshold1=lower , threshold2=upper+100)
plt.imshow(edges)
```

```python
edges = cv2.Canny(image=img, threshold1=lower+100 , threshold2=upper)
plt.imshow(edges)
```

```python
edges = cv2.Canny(image=img, threshold1=lower+100 , threshold2=upper+100)
plt.imshow(edges)
```

```python
edges = cv2.Canny(image=img, threshold1=lower+100 , threshold2=upper+150)
plt.imshow(edges)
```

```python
edges = cv2.Canny(image=img, threshold1=lower+150 , threshold2=upper+175)
plt.imshow(edges)
```

 # Manual Blurring before Edge Detection


```python
for k in [3,5,7,9,11]:
    blurred_img = cv2.blur(img,ksize=(k,k))
    edges = cv2.Canny(image=blurred_img, threshold1=lower , threshold2=upper)
    plt.title(f"{k}x{k}")
    plt.imshow(edges)
    
    plt.show()
```
