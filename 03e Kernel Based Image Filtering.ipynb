```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def load_img():
    img = cv2.imread('DATA/00-puppy.jpg', cv2.IMREAD_GRAYSCALE)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return img
img = load_img() 
```

```python
def display_img(img, title):
    fig = plt.figure(figsize=(6,5))
    ax = fig.add_subplot(111)
    ax.set_title(title)
    ax.imshow(img, cmap='gray')
    plt.show()
```

```python
plt.imshow(img, cmap='gray')
```

```python
def convolve(image, kernel):
    return cv2.filter2D(image, -1, kernel)
```

```python
kernels = {
    'identity': np.array([
        [0, 0, 0],
        [0, 1, 0],
        [0, 0, 0]
    ]),
    'blur': np.array([
        [0.0625, 0.125, 0.0625],
        [0.125, 0.25, 0.125],
        [0.0625, 0.125, 0.0625],
    ]),
    'top_sobel': np.array([
        [1, 2, 1],
        [0, 0, 0],
        [-1, -2, -1]
    ]),
    'bottom_sobel': np.array([
        [-1, -2, -1],
        [0, 0, 0],
        [1, 2, 1]
    ]),
    'left_sobel': np.array([
        [1, 0, -1],
        [2, 0, -2],
        [1, 0, -1]
    ]),
    'right_sobel': np.array([
        [-1, 0, 1],
        [-2, 0, 2],
        [-1, 0, 1]
    ]),
    'outline': np.array([
        [-1, -1, -1],
        [-1, 8, -1],
        [-1, -1, -1]
    ]),
    'sharpen': np.array([
        [0, -1, 0],
        [-1, 5, -1],
        [0, -1, 0]
    ]),
    'emboss1': np.array([
        [-2, -1, 0],
        [-1, 1, 1],
        [0, 1, 2]
    ]),
    'emboss2': np.array([
        [2, 1, 0],
        [1, 1, -1],
        [0, -1, -2]
    ]),
    'emboss3': np.array([
        [0, -1, -2],
        [1, 1, -1],
        [2, 1, 0]
    ]),
    'emboss4': np.array([
        [0, 1, 2],
        [-1, 1, 1],
        [-2, -1, 0]
    ]),
}
```

```python
for k, v in kernels.items():
    filtered_img = convolve(img, v)
    display_img(filtered_img, k)
```
