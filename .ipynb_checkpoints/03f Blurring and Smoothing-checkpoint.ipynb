```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def load_img():
    img = cv2.imread('DATA/00-puppy.jpg', cv2.IMREAD_GRAYSCALE)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return img
img = load_img()    
```

```python
def display_img(img, title):
    fig = plt.figure(figsize=(6,5))
    ax = fig.add_subplot(111)
    ax.set_title(title)
    ax.imshow(img, cmap='gray')
    plt.show()
```

```python
plt.imshow(img, cmap='gray')
```

# Mean Filters

**MANUAL FILTERS**

```python
def convolve(image, kernel):
    return cv2.filter2D(image, -1, kernel)
```

```python
kernels = {
    '3x3': np.ones(shape=(3,3), dtype=np.float32)/9,
    '5x5': np.ones(shape=(5,5), dtype=np.float32)/25,
    '7x7': np.ones(shape=(7,7), dtype=np.float32)/48
}
```

```python
for k, v in kernels.items():
    filtered_img = convolve(img, v)
    display_img(filtered_img, f"{k}x{k}")
```

**CV2 FILTERS**

```python
for k in [3,5,7]:
    filtered_img = cv2.blur(img, (k, k))
    display_img(filtered_img, f"{k}x{k}")
```

# Gaussian Filter

```python
for k in [3,5,7]:
    filtered_img = cv2.GaussianBlur(img, (k, k), 10)
    display_img(filtered_img, f"{k}x{k}")
```

# Median Blur


```python
for k in [3,5,7]:
    filtered_img = cv2.medianBlur(img, k)
    display_img(filtered_img, f"{k}x{k}")
```

# Bilateral Filtering

```python
for k in [3,5,7]:
    filtered_img = cv2.bilateralFilter(img, k, 75, 75)
    display_img(filtered_img, f"{k}x{k}")
```
