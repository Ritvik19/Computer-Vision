# Class Based Regex

```python
from sklearn.metrics import confusion_matrix
```

**DIGIT**

```python
DIGIT_REGEX = re.compile(r'''(
                            ((\d)+
                            (-)?(,)?(\s)?
                            )|(
                            (\d\s)+
                            (\d))
                                )''', re.VERBOSE)
```

```python
YEAR_REGEX = re.compile(r'((1\d\d\d)|(20\d\d))(,)?(\s)?')
```

```python
YEAR_REGEX.fullmatch('1958, ')
```

```python
def match_digit(x):
    if YEAR_REGEX.fullmatch(str(x)):
        return 0
    else:
        if DIGIT_REGEX.fullmatch(str(x)):
            return 1
        else:
            return 0
```

```python
DIGIT_REGEX.fullmatch(str('47,'))
```

```python
digit_matched = data_['before'].apply(match_digit)

print(confusion_matrix(data_['DIGIT'], digit_matched))
```
